## Extract files
Please remove any existing file path before proceeding. It won't delete it for you without permission.

### copy subset only

```python
# load and extract
import os
import random
import shutil

image_dir = dataset_dir + "/images"
label_dir = dataset_dir + "/labels"

# Get all image files (adjust extensions if needed)
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]

# Set seed for reproducibility
random.seed(random_seed)
sample_files = random.sample(image_files, num_samples)

# copy subset sample to "images/" and "labels/""
output_image_dir = output_dir + "/images"
output_label_dir = output_dir + "/labels"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)
for file in sample_files:    
    shutil.copy(os.path.join(image_dir, file), os.path.join(output_image_dir, file))
    label_file = file.rsplit('.', 1)[0] + '.txt'
    shutil.copy(os.path.join(label_dir, label_file), os.path.join(output_label_dir, label_file))
    
    # print(os.path.join(image_dir, "->", file), os.path.join(output_image_dir, file))
    # label_file = file.rsplit('.', 1)[0] + '.txt'
    # print(os.path.join(label_dir, "->", label_file), os.path.join(output_label_dir, label_file))
    

# contruct YAML config file
data_yaml_path = output_dir + '/data.yaml'
classes = ['hole', 'pole', 'stairs', 'bottle', 'rock']
with open(data_yaml_path, 'w') as f: 
    f.write(f"train: ./images\n")   # Not used
    f.write(f"val: ./images\n")     # Not used
    f.write(f"test: ./images\n")    # Used
    f.write(f"nc: {len(classes)}\n")
    f.write(f"names: {classes}\n")
    
print("YAML file created at: ", data_yaml_path)
```

### copy subset to 'eval' and remaining to 'base'
Selected sample move to eval, the rest for base

```python
# load and extract
import os
import random
import shutil

image_dir = dataset_dir + "/images"
label_dir = dataset_dir + "/labels"

# Get all image files (adjust extensions if needed)
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]

# Set seed for reproducibility
random.seed(random_seed)
sample_files = random.sample(image_files, num_samples)

# copy subset sample to "eval/images/" and "eval/labels/""
output_image_dir = output_dir + "/eval/images"
output_label_dir = output_dir + "/eval/labels"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)
for file in sample_files:    
    shutil.copy(os.path.join(image_dir, file), os.path.join(output_image_dir, file))
    label_file = file.rsplit('.', 1)[0] + '.txt'
    shutil.copy(os.path.join(label_dir, label_file), os.path.join(output_label_dir, label_file))

# contruct YAML config file
data_yaml_path = output_dir + '/eval/data.yaml'
classes = ['hole', 'pole', 'stairs', 'bottle', 'rock']
with open(data_yaml_path, 'w') as f: 
    f.write(f"train: ./images\n")   # Not used
    f.write(f"val: ./images\n")     # Not used
    f.write(f"test: ./images\n")    # Used
    f.write(f"nc: {len(classes)}\n")
    f.write(f"names: {classes}\n")
    
print("YAML file created at: ", data_yaml_path)

# copy remaining to base/images/ and base/lables
output_image_dir = output_dir + "/base/images"
output_label_dir = output_dir + "/base/labels"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)
for file in image_files:    
    if file not in sample_files:
        shutil.copy(os.path.join(image_dir, file), os.path.join(output_image_dir, file))
        label_file = file.rsplit('.', 1)[0] + '.txt'
        shutil.copy(os.path.join(label_dir, label_file), os.path.join(output_label_dir, label_file))

# contruct YAML config file
data_yaml_path = output_dir + '/base/data.yaml'
classes = ['hole', 'pole', 'stairs', 'bottle', 'rock']
with open(data_yaml_path, 'w') as f: 
    f.write(f"train: ./images\n")   # Not used
    f.write(f"val: ./images\n")     # Not used
    f.write(f"test: ./images\n")    # Used
    f.write(f"nc: {len(classes)}\n")
    f.write(f"names: {classes}\n")
    
print("YAML file created at: ", data_yaml_path)
```

## Compress dataset from a folder

```python
# zip to a file for upload to a remote storage
import zipfile
import os
zip_dir = input("Enter dataset folder location: ")
print(zip_dir)

zip_filename = input("Enter ZIP file name: ")
print(zip_filename)

output_image_dir = zip_dir + "/images"
output_label_dir = zip_dir + "/labels"


zip_filename = zip_dir + "/" + zip_filename

print("Saving zip file to: ", zip_filename)

with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for folder in [output_image_dir, output_label_dir]:
        for root, dirs, files in os.walk(folder):
            for file in files:
                filepath = os.path.join(root, file)
                arcname = os.path.relpath(filepath, start=os.path.dirname(folder))
                zipf.write(filepath, arcname)
    
    filepath = os.path.join(zip_dir, "data.yaml")
    arcname = os.path.relpath(filepath, start=os.path.dirname(folder))
    zipf.write(filepath, arcname)
                
print("Done")
```
