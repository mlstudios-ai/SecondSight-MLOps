# EnigmaAI MLOps

To allow appropriate compute resources allocation (e.g different worker resources) and preventing bottleneck, use the following queue names for different types of tasks when starting a ClearML Agent.

## Non GPU tasks - use **`default`** queue

- Copying files

- Download/upload

- Visualisation

- Performance metrics display

## GPU tasks queues

- **'processing`** for hazarded detection processing tasks

- **`training`** for hazard detection model training

- **`deployment`** for evaluation, validation, and deployment

- **`hd_serving`** for serving endpoints for hazard detection

- **`sd_serving`** for serving endpoints for scene description

```python
import clearml
```

```python
# Running from local, using .env from root project folder
import os
from dotenv import load_dotenv
load_dotenv()

from clearml import Task
web_server = os.getenv("CLEARML_WEB_HOST")
api_server = os.getenv("CLEARML_API_HOST")
files_server = os.getenv("CLEARML_FILES_HOST")
access_key = os.getenv("CLEARML_API_ACCESS_KEY")
secret_key = os.getenv("CLEARML_API_SECRET_KEY")

Task.set_credentials(web_host=web_server,
                     api_host=api_server,
                     files_host=files_server,
                     key=access_key,
                     secret=secret_key
                     )
```
