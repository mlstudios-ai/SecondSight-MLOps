# EnigmaAI MLOps

To allow appropriate compute resources allocation (e.g different worker resources) and preventing bottleneck, use the following queue names for different types of tasks when starting a ClearML Agent.

## Non GPU tasks - use **`default`** queue

- Copying files

- Download/upload

- Visualisation

- Performance metrics display

## GPU tasks queues

- **'processing`** for hazarded detection processing tasks

- **`training`** for hazard detection model training

- **`deployment`** for evaluation, validation, and deployment

- **`hd_serving`** for serving endpoints for hazard detection

- **`sd_serving`** for serving endpoints for scene description
